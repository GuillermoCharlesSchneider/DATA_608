```python
url = "https://raw.githubusercontent.com/python-visualization/folium/main/examples/data/us-states.json"
state_geo = requests.get(url).json()
```

```python
total_map = dict(zip(state_data["States"], state_data["Total Production"]))
consum_map = dict(zip(state_data["States"], state_data["Consumption"]))
diff_map = dict(zip(state_data["States"], state_data["Difference"]))
coal_map   = dict(zip(state_data["States"], state_data["Coal"]))
gas_map   = dict(zip(state_data["States"], state_data["Natural Gas"]))
oil_map   = dict(zip(state_data["States"], state_data["Crude Oil"]))
nuclear_map = dict(zip(state_data["States"], state_data["Nuclear"]))
renew_map = dict(zip(state_data["States"], state_data["Renewable"]))

for feature in state_geo["features"]:
    state = feature["properties"]["name"]
    feature["properties"]["Total Production"] = total_map.get(state)
    feature["properties"]["Consumption"] = consum_map.get(state)
    feature["properties"]["Difference"] = diff_map.get(state)
    feature["properties"]["Coal"] = coal_map.get(state)
    feature["properties"]["Natural Gas"] = gas_map.get(state)
    feature["properties"]["Crude Oil"] = oil_map.get(state)
    feature["properties"]["Nuclear"] = nuclear_map.get(state)
    feature["properties"]["Renewable"] = renew_map.get(state)
    
```
